```python
import pandas as pd
import numpy as np
```

```python
%cd "/content/drive/MyDrive/My-body-Your-choice/data/entailment_labled_GPT"
```

```python
data = pd.read_excel('unique_all_topics_enatilment_labled_by_GPT4_with_keyword_abortion.xlsx',header=41)
data.head(3)
```

```python
def preprocess_label(text):
    if text == 'Forced':
        return 1
    else:
        return 0
```

```python
data['Forced_by_anyone'] = data['Manual_label'].apply(preprocess_label)
```

```python
data['Forced_by_husband'] = np.where((data['Manual_label'] == 'Forced') & (data['Who'].isna()), 1, 0)
```
