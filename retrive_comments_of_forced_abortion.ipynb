```python
all_comments = pd.read_csv("data/comments_on_topics_with_keyword_abortion(سقط).csv")
```

```python
topics = pd.read_csv("data/entailment_labled_GPT/unique_all_topics_enatilment_labled_by_GPT4_with_keyword_abortion_with_forced_label.csv")
```

```python
selected_topics = topics.loc[topics.Forced_by_husband == 1]
selected_topics
```

```python
selected_commenst = all_comments.loc[all_comments.topic_id.isin(selected_topics.topic_id)]
```

```python
selected_commenst.to_csv("data/comments_analyising//comments_on_topics_with_keyword_abortion(سقط)_with_forced_by_husban_label.csv")
```

```python
print(selected_commenst.shape)
selected_topics.rename(columns={"userid": "topic_author_id"},inplace=True)
selected_topics = selected_topics[["topic_id","topic_author_id","Forced_by_husband"]]
selected_commenst.merge(selected_topics,on="topic_id",how="inner")
selected_commenst.drop_duplicates(subset="id_x",inplace=True)
print(selected_commenst.shape)
selected_commenst
```
