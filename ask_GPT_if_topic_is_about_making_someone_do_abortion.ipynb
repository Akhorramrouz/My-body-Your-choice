```python
import pandas as pd
import time
import openai

# Set your OpenAI API key here
openai.api_key = 'API_key'

def text_entailment(hypothesis, premise):
    """
    Analyzes the logical relationship between two Persian sentences.
    Returns the type of relationship and an explanation.
    """
    prompt = f"""Please analyze the logical relationship between the following Persian sentences. Do they entail each other, are they neutral, or do they contradict each other?
premise = "{premise}"
hypothesis = "{hypothesis}"

answer with only one of the following: entailment, neutral, contradiction

then after ### explain the reason.
"""
    try:
        response = openai.ChatCompletion.create(
            model='gpt-4',
            messages=[
                {'role': 'system', 'content': 'you are an Iranian social scientist.'},
                {'role': 'user', 'content': prompt}
            ]
        )
        return response.choices[0].message['content']
    except Exception as e:
        print(f"Error in text_entailment: {e}")
        return "error"

def process_data(data, start_index, step_index, hypothesis):
    for start in range(start_index, len(data), step_index):
        end_index = start + step_index
        print(start, end_index)

        sampled_data = data.iloc[start:end_index].copy()
        sampled_data['being_force_for_abortion'] = [text_entailment(hypothesis, text) for text in sampled_data['text']]
        sampled_data.to_csv(f'data/entailment_labled_GPT/abortion_entailment_{start}_{end_index}.csv')
        time.sleep(5)

# Load data and process
data = pd.read_csv('data/topics_with_keyword_abortion(سقط)_entailment_shuffled_all.csv', index_col=0)
hypothesis = "someone is pressured or suggested or forced into an abortion."

```

```python
process_data(data, start_index=680, step_index=5, hypothesis=hypothesis)
```
